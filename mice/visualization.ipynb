```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# snout coordinates from training data
data = pd.read_csv("data/CollectedData_Pranav.csv", header=2).iloc[:,1:3].astype(float)
data['y'] *= -1  # y-axis in label csv is 'flip' of normal cartesian coordinates
data
```

```python
# activity heatmap using 2d KDE

fig, ax = plt.subplots()
sns.kdeplot(data=data, x='x', y='y', fill=True, ax=ax, bw_adjust=5e-1, cut=3, levels=100);
ax.set_xlim(0, 640)
ax.set_ylim(-480, 0)
plt.show()
```

```python
# movement trajectory

fig, ax = plt.subplots()
sns.lineplot(data=data, x='x', y='y', sort=False, lw=1, ax=ax, ci=None);
ax.set_xlim(0, 640)
ax.set_ylim(-480, 0)
plt.show()
```
